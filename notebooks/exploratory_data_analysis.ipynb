```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import yfinance as yf
```

```python
nav = pd.read_csv("../data/processed/nav_clean.csv")
nav['Indexed'] = (nav['Total'] / nav['Total'].iloc[0]) * 100
nav['ReportDate'] = pd.to_datetime(nav['ReportDate'])

sns.set_theme(style="ticks", font='serif', font_scale=1.1)
fig, ax = plt.subplots(figsize=(16, 8), dpi=600)
sns.lineplot(data=nav, x="ReportDate", y="Indexed")
plt.rcParams['font.serif'] = ['Minion Pro', 'Times New Roman']
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b.'))
ax.margins(x=0, y=0)
sns.despine()
plt.xticks(rotation=0, ha='left')
plt.xlabel(xlabel="Date")
plt.ylabel(ylabel="Performance in %")
plt.tight_layout()
plt.savefig('../figures/performance.png', 
            dpi=600,
            bbox_inches='tight',
            facecolor='white')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# trades
trades = pd.read_csv('../data/raw/ibkr_data.csv', skiprows=271)
trades = trades.drop(columns="Taxes")
trades['DateTime'] = pd.to_datetime(
    trades['DateTime'].str.replace(';', ''),
    format='%Y%m%d%H%M%S'
)
#trades.head(50)
#trades.to_csv('../data/processed/trades_clean.csv', index=False)
```
